```python
import numpy as np
import pandas as pd
import sympy 
import matplotlib.pyplot as plt
import sys
from scipy import fftpack
from scipy.fftpack import fft
from scipy import signal
from scipy.fft import fftshift
```

```python
df = pd.read_csv("C:/Users/YASHWANT MOHARIL/Downloads/111_Healthy_13.csv")
```

```python
df.head
```

```python
x = df.iloc[:,:].values
t = np.linspace(start = 0, stop = 0.15, num = 2999)
print(t)
t.size
```

```python
# sampling rate
sr = 20000
# sampling interval
X = fft(df)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T
f = np.array(freq)
f.resize(2999,1)
print(f)
```

```python
plt.plot(f, np.abs(X), color = 'red')
plt.title('Frequency Domain')
plt.xlabel('Frequency')
plt.ylabel('Numeric Data')
plt.show()
```

```python
x = x.flatten()
plt.specgram(x, Fs = sr, cmap = 'jet')
plt.title('Signal Spectogram')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
```

```python
f, T, Zxx = signal.stft(x, sr)
plt.pcolormesh(T, f, np.abs(Zxx), cmap = 'jet', shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

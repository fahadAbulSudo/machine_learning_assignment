```python
import numpy as np
import os
import pandas as pd
import sympy
import matplotlib.pyplot as plt
import sys
from scipy.fftpack import rfft
from scipy.fft import rfft,rfftfreq
from scipy.fft import fft, ifft#
from scipy import signal
```

```python
path = r'2nd_test'
filename = '2004.02.19.05.32.39'
dataset=pd.read_csv(os.path.join(path, filename), sep='\t',header=None)
dataset=pd.read_csv(r'2nd_test/2004.02.19.05.32.39', sep='\t',header=None)
```

```python
dataset
```

```python
for i in [0,1,2,3]:
    x = dataset.iloc[:,i].values
    # sampling rate
    sr = 20480
    # sampling interval
    X = rfft(x)#takes only unique values in dataset
    N = len(X)
    n = np.arange(N)#NumPy arange() is one of the array creation routines based on numerical ranges. It creates an instance of ndarray with evenly spaced values and returns the reference to it
    T = N/sr
    freq = n/T
    f = np.array(freq)
    np.set_printoptions(precision=8,suppress=True)
    len(f)
    plt.plot(f, np.abs(X), color = 'blue')
    plt.title("Frequency Domain")
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.show()
```
